# Linear Regression using PyTorch

```python
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('seaborn')
```

```python
X = pd.read_csv("../Linear Regression/Linear_X_Train.csv").values
y = pd.read_csv("../Linear Regression/Linear_Y_Train.csv").values
```

```python
plt.scatter(X,y,color="green")
plt.show()
```

```python
X, y = tr.tensor(X),tr.tensor(y)
```

```python
m = tr.nn.Parameter(data = tr.randn(1))
c = tr.nn.Parameter(data = tr.randn(1))

y_pred = m*X+c
```

```python
plt.scatter(X.numpy(),y.numpy(),color="green")
plt.plot(X,y_pred.detach().numpy())
plt.show()
```

```python
def mse(y,y_pred):
    return ((y-y_pred)**2).mean()
loss_list=[]
for i in range(200):
    y_pred = m*X+c
    lr = 0.1
    loss = mse(y,y_pred)
    loss_list.append(loss)
    loss.backward()

    with tr.no_grad():
        m -= lr*m.grad
        c -= lr*c.grad

        m.grad.zero_()
        c.grad.zero_()
```

```python
plt.scatter(X.numpy(),y.numpy(),color="green")
plt.plot(X,y_pred.detach().numpy())
plt.show()
```

```python
plt.plot(loss_list)
```
